```python
api_key = os.getenv("GOOGLE_API_KEY")
```

```python
client = genai.configure(api_key=api_key)
```

```python
MODEL_NAME = "gemini-2.5-flash"
```

```python
model = genai.GenerativeModel(MODEL_NAME)
```

```python
prompt = """
Identify all the data and it's key in the image.

1. Net Cash Flow from Operating Activities
2. Net Cash Flow from Investing Activities
3. Net Cash Flow from Financing Activities
4. Net Increase / Decrease in Cash
5. Opening Balance
6. Closing Balance

Return every data in JSON
"""
```

```python
img = Image.open("cashflow_image.jpeg")
```

```python
type(img)
```

```python
response = model.generate_content(
    contents=[img, prompt],
    generation_config={
        "temperature":0.2,
        "max_output_tokens":5000
    }
)
```

```python
from google import genai
```

```python
client = genai.Client(api_key=api_key)
```

```python
import cv2

img = cv2.imread("cashflow_image.jpeg")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

thresh = cv2.adaptiveThreshold(
    gray,255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    11,2
)

cv2.imwrite("processed.png", thresh)
```

```python
print("List of models that support generateContent:\n")
for m in client.models.list():
    for action in m.supported_actions:
        if action == "generateContent":
            print(m.name)    
```

```python
def ask_llm(prompt):
    response = client.models.generate_content(
        
        model = "gemini-2.5-flash",
        contents=[img, prompt]
    )
    return response
```

```python
output=ask_llm("What is DPDP act?")
print(output)
```
